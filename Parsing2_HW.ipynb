```python
# импортируем библиотеки
from pprint import pprint
from lxml import html
import requests
```

```python
url = 'https://lenta.ru/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36',
}
# прикрываемся сайтом
response = requests.get(url, headers=headers)

response  # получим статус ответа
```

```python
dom = html.fromstring(response.text)
```

```python
news = dom.xpath('//div/a[contains(@class, "card")]')
# не парсим весь сайт, а находим "блоки" с нужным содержимым
```

```python
len(news)
```

```python
news_list = []

for new in news:
  news_name = new.xpath('./div/*/text()')[0]
  news_link = new.xpath('./@href')[0]
  if 'http' not in news_link:
    news_link = url + news_link
  pub_time = new.xpath('./div/div/time/text()')

  news_dict = {
      'name': news_name,
      'link': news_link,
      'publication_time': pub_time,
  }


  
  news_list.append(news_dict)

news_list
```
